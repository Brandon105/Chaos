```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma = 16
b = 4
r = 45.92

t = np.arange(0, 15, .002)
X0, Y0, Z0 = 1, 5, 10
f0 = X0, Y0, Z0

def deriv(f, t, sigma, b, r):
    X,Y,Z = f
    dXdt = sigma*(Y - X)
    dYdt = r*X - Y - X*Z
    dZdt = X*Y - b*Z
    return dXdt, dYdt, dZdt
ret = odeint(deriv, f0, t, args=(sigma, b, r))
X, Y, Z = ret.T
```

```python
plt.plot(t, X, label = 'X(t)')
plt.plot(t, Y, label = 'Y(t)')
plt.plot(t, Z, label = 'Z(t)')
plt.legend()
plt.show()
```

```python
def deriv2(f2, t, sigma, b, r):
    X, Y, Z, X2, Z2 = f2
    dX2dt = sigma*(Y - X2)
    dZ2dt = X2*Y - b*Z2
    dXdt = sigma*(Y - X)
    dYdt = r*X - Y - X*Z
    dZdt = X*Y - b*Z
    return dX2dt, dZ2dt, dXdt, dYdt, dZdt
ret2 = odeint(deriv2, f0, t, args=(sigma, b, r))
X, Y, Z, X2, Z2 = ret2.T
```
